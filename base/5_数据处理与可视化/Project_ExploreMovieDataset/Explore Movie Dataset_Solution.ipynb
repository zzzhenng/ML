**任务3.2：**分析电影净利润（票房-成本）随着年份变化的情况，并简单进行分析。

```python
movie_data['profit'] = movie_data['revenue'] - movie_data['budget']
profit = movie_data.groupby(['release_year'])['profit']
index = profit.mean().index

plt.figure(figsize=(12, 12))

plt.subplot(3,1,1)
sb.pointplot(index, profit.mean())
plt.xticks(rotation=90)
plt.xlabel('release year')
plt.ylabel('profit')
plt.title('mean profit')
plt.tight_layout()

plt.subplot(3,1,2)
sb.pointplot(index, profit.sum())
plt.xticks(rotation=90)
plt.xlabel('release year')
plt.ylabel('profit')
plt.title('sum profit')
plt.tight_layout()

plt.subplot(3,1,3)
sb.pointplot(index, profit.count())
plt.xticks(rotation=90)
plt.xlabel('release year')
plt.ylabel('number')
plt.title('release num')
plt.tight_layout()
```

分析上图，结论如下：
- 电影的净利润从1960~1990年间基本保持增长，其中1977年增长最大。1990年之后电影的净利润趋于稳定。
- 电影的总利润从1960~1990年间增速平缓，之后增速变大。
- 每年的电影的发行数与总利润增长趋势接近。

**[选做]任务3.3：**选择最多产的10位导演（电影数量最多的），绘制他们排行前3的三部电影的票房情况，并简要进行分析。

```python
# 1. 选择最多产的10位导演
director10 = movie_data['director'].value_counts()[0:10]
# 2. 获得每位导演对应的Top3票房电影
plt.figure(figsize=(20,30))

revenues = []
for i in range(10):
    d = director10.index[i]
    d_movie = movie_data[movie_data['director'] == d]
    revenue = d_movie.sort_values('revenue', ascending=False)[0:3]
    revenues.append(revenue)
    
    plt.subplot(2,5,i+1)
#     plt.bar(revenue['original_title'], revenue['revenue'])
    sb.barplot(revenue['original_title'], revenue['revenue'], color=sb.color_palette()[0])
    plt.xticks(rotation=15)
    plt.ylim([0, 1.2e9])
    plt.title(d)
```

分析上图，结论如下：
- 最高产的导演票房不一定最高，仅靠数量是不够的。
- Steve Speilberg导演不仅高产，且票房最高。

**[选做]任务3.4：**分析1968年~2015年六月电影的数量的变化。

```python
# 1. 提取1968~2015年
sel_year = movie_data['release_year'].isin(list(range(1968, 2016)))
# 2. 提取六月
month = list(map(lambda x: x.split('/')[0], movie_data['release_date']))
sel_june = list(map(lambda x: x=='6', month))
# 3. 组合
movie_june = movie_data[sel_year & sel_june]
# 4. 作图
movie_june['release_year'].value_counts().sort_index().plot(kind='bar', figsize=(15, 5), color = 'blue');
plt.xlabel('Release Year');
plt.ylabel('Movie Count');
```

---

**[选做]任务3.5：**分析1968年~2015年六月电影 `Comedy` 和 `Drama` 两类电影的数量的变化。

```python
movie_genres = movie_june[movie_june.genres.isin(['Comedy', 'Drama'])]

plt.figure(figsize = [15, 6])
sb.countplot(data = movie_genres, x='release_year', hue='genres', palette="Set2");
plt.xticks(rotation = 90)
plt.xlabel('Release Year');
plt.ylabel('Movie Count');
```
